```python
#removes grammer words
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords',quiet=True)
stop_words=set(stopwords.words('english'))
text="Today iam getting  a bike"
tokens=[]
for word in text.split():
  if word not in stop_words:
    tokens.append(word)
print(tokens)
```

```python
#seperates the words along with inverted commas
import nltk
nltk.download('punkt_tab',quiet=True)
sentence="I haven't  seen the movie"
basic=sentence.split()
print('Basic split:',basic)
from nltk.tokenize import word_tokenize
smart=word_tokenize(sentence.lower())
print('NLTK tokenize:',smart)
```

```python
import nltk
from nltk.stem import PorterStemmer,WordNetLemmatizer
nltk.download('wordnet',quiet=True)
stemmer=PorterStemmer()
lemmatizer=WordNetLemmatizer()
words=['running','studies','better']
for word in words:
  stem=stemmer.stem(word)
  lemma=lemmatizer.lemmatize(word,pos='v')
  print(f'{word:12}|stem:{stem:10}|lemma:{lemma}')
```

```python
import re
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
nltk.download(['stopwords','wordnet','punkt'],quiet=True)
stop_words=set(stopwords.words('english'))
lemmatizer=WordNetLemmatizer()
def preprocess(text):
  text=text.lower()
  text=re.sub(r'[^a-z]',"",text)
  tokens=text.split()
  filtered_tokens=[]
  for w in tokens:
    if w not in stop_words:
      filtered_tokens.append(w)
  tokens=filtered_tokens
  lemmatized_tokens=[]
  for w in tokens:
    lemmatized_tokens.append(lemmatizer.lemmatize(w))
  tokens=lemmatized_tokens
  return' '.join(tokens)

print(preprocess('I am waiting in park'))
```

```python
from sklearn.feature_extraction.text import CountVectorizer
emails=[
    'cricket match cricket match cricket',
    'free win money free win money win',
    'cricket match free win money cricket',
]
cv=CountVectorizer()
x=cv.fit_transform(emails).toarray()
print('vocabulary:',cv.get_feature_names_out())
print('Row Matrix:')
print(x)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
emails=[
    'cricket match cricket match cricket',
    'free win money free win money win',
    'cricket match free win money cricket',
]
tfidf=TfidfVectorizer(sublinear_tf=True,max_df=1.0,min_df=1)
x_tfidf=tfidf.fit_transform(emails).toarray()
print('Feature names:',tfidf.get_feature_names_out())
print('TF-IDF matrix(round)')
print(x_tfidf)
```
